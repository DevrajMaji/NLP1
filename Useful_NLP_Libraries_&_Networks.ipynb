Question 1: Compare and contrast NLTK and spaCy in terms of features, ease of use,and performance.

->NLTK and spaCy are popular libraries used for Natural Language Processing (NLP) in Python.

NLTK provides many tools for tasks like tokenization, stemming, lemmatization, and text analysis. It also includes many datasets and linguistic resources, which makes it very useful for learning and research in NLP. However, it often requires more code and manual setup.

spaCy, on the other hand, is designed for real-world applications. It provides built-in features such as part-of-speech tagging, named entity recognition, and dependency parsing. It also includes pre-trained models that make NLP tasks easier to implement.

In terms of performance, spaCy is faster and more efficient than NLTK because it is optimized for processing large amounts of text. Overall, NLTK is better for learning and experimentation, while spaCy is better for building production-level NLP applications.

Question 4: Describe the architecture and functioning of a Recurrent Natural Network(RNN).

->A Recurrent Neural Network (RNN) is a type of neural network designed to work with sequential data such as text, speech, or time-series data.

Architecture:
An RNN consists of an input layer, hidden layer, and output layer. Unlike traditional neural networks, the hidden layer in an RNN has a loop connection, which means the output from the previous step is fed back into the network. This allows the model to remember previous information in the sequence.

Functioning:
When a sequence of data is given as input, the RNN processes it one step at a time. At each step, it takes the current input and the previous hidden state to produce a new hidden state and output. This process helps the network capture dependencies between past and present data.

Because of this memory capability, RNNs are widely used in tasks like language modeling, speech recognition, machine translation, and text generation.


Question 5: What is the key difference between LSTM and GRU networks in NLP
applications?

->The main difference between LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) networks is in their structure and complexity.

LSTM has three gates: input gate, forget gate, and output gate, along with a separate cell state to store long-term information. This structure helps it remember information for a long time but makes the model more complex.

GRU, on the other hand, has only two gates: update gate and reset gate, and it combines the hidden state and cell state into one. Because of this simpler structure, GRU is faster to train and uses fewer parameters than LSTM.

In NLP applications, both are used to handle sequential data, but LSTM is better for capturing very long dependencies, while GRU is often preferred when faster training and simpler models are needed.



```python
'''Question 6: Write a Python program using TextBlob to perform sentiment analysis on
the following paragraph of text:
вҖңI had a great experience using the new mobile banking app. The interface is intuitive,
and customer support was quick to resolve my issue. However, the app did crash once
during a transaction, which was frustrating"
Your program should print out the polarity and subjectivity scores'''

from textblob import TextBlob

text = """I had a great experience using the new mobile banking app.
The interface is intuitive, and customer support was quick to resolve my issue.
However, the app did crash once during a transaction, which was frustrating."""

blob = TextBlob(text)

sentiment = blob.sentiment

print("Polarity:", sentiment.polarity)
print("Subjectivity:", sentiment.subjectivity)
```

```python
'''Question 7: Given the sample paragraph below, perform string tokenization and
frequency distribution using Python and NLTK:
вҖңNatural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical.вҖқ
'''


import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

tokens = word_tokenize(text)

print("Tokens:")
print(tokens)

freq = FreqDist(tokens)

print("\nFrequency Distribution:")
print(freq)

print("\nMost Common Words:")
print(freq.most_common(10))
```

```python
'''Question 8: Implement a basic LSTM model in Keras for a text classification task using
the following dummy dataset. Your model should classify sentences as either positive
(1) or negative (0).
# Dataset
texts = [
вҖңI love this projectвҖқ, #Positive
вҖңThis is an amazing experienceвҖқ, #Positive
вҖңI hate waiting in lineвҖқ, #Negative
вҖңThis is the worst serviceвҖқ, #Negative
вҖңAbsolutely fantastic!вҖқ #Positive
]
labels = [1, 1, 0, 0, 1]
Preprocess the text, tokenize it, pad sequences, and build an LSTM model to train on
this data. You may use Keras with TensorFlow backend.
'''

import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Dataset
texts = [
"I love this project",
"This is an amazing experience",
"I hate waiting in line",
"This is the worst service",
"Absolutely fantastic!"
]

labels = [1, 1, 0, 0, 1]


tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

max_len = 5
X = pad_sequences(sequences, maxlen=max_len)

y = np.array(labels)

model = Sequential()
model.add(Embedding(input_dim=50, output_dim=16, input_length=max_len))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X, y, epochs=10, verbose=1)

model.summary()
```

```python
'''Question 9: Using spaCy, build a simple NLP pipeline that includes tokenization,
lemmatization, and entity recognition. Use the following paragraph as your dataset:
вҖңHomi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of IndiaвҖҷs atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India.вҖқ
Write a Python program that processes this text using spaCy, then prints tokens, their
lemmas, and any named entities found.'''


import spacy


nlp = spacy.load("en_core_web_sm")

text = """Homi Jehangir Bhaba was an Indian nuclear physicist who played a key role in the
development of IndiaвҖҷs atomic energy program. He was the founding director of the Tata
Institute of Fundamental Research (TIFR) and was instrumental in establishing the
Atomic Energy Commission of India."""


doc = nlp(text)

print("Tokens and Lemmas:")
for token in doc:
    print(token.text, "->", token.lemma_)


print("\nNamed Entities:")
for ent in doc.ents:
    print(ent.text, "-", ent.label_)
```

```python
'''Question 10: You are working on a chatbot for a mental health platform. Explain how
you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford
NLP to understand and respond to user input effectively. Detail your architecture, data
preprocessing pipeline, and any ethical considerations.'''


'''
1. Architecture

The chatbot system would have the following components:

User Input Layer вҖ“ The user sends a message describing their feelings or problems.

NLP Processing Layer вҖ“ Libraries like spaCy or Stanford NLP perform tasks such as tokenization, lemmatization, and entity recognition.

Text Vectorization вҖ“ The processed text is converted into numerical form using tokenization and padding.

Deep Learning Model (LSTM/GRU) вҖ“ The model analyzes the sequence of words and predicts the intent or emotional state of the user.

Response Generator вҖ“ Based on the predicted intent (e.g., stress, anxiety, sadness), the chatbot returns an appropriate supportive response.

LSTM or GRU networks are useful because they can understand context and word order, which is important when analyzing emotional conversations.

2. Data Preprocessing Pipeline

The preprocessing steps include:

Tokenization вҖ“ Splitting sentences into words.

Lowercasing вҖ“ Converting text to lowercase.

Lemmatization вҖ“ Converting words to their base form.

Stopword removal (optional) вҖ“ Removing common words that do not add meaning.

Tokenization and padding вҖ“ Converting text into numerical sequences and ensuring equal length.

3. Ethical Considerations

When developing a mental health chatbot, several ethical issues must be considered:

Privacy and data security вҖ“ User conversations must be kept confidential.

Bias prevention вҖ“ Training data should be diverse to avoid biased responses.

Transparency вҖ“ Users should know they are interacting with an AI system.

Safety вҖ“ The chatbot should recommend professional help if it detects severe distress or crisis situations.'''

import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Sample dataset
texts = [
"I feel very sad today",
"I am happy and relaxed",
"I am feeling anxious about my exams",
"I feel calm and peaceful"
]

labels = [0, 1, 0, 1]   # 0 = negative emotion, 1 = positive emotion

# Preprocessing with spaCy
processed_texts = []
for text in texts:
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc]
    processed_texts.append(" ".join(tokens))

tokenizer = Tokenizer()
tokenizer.fit_on_texts(processed_texts)
sequences = tokenizer.texts_to_sequences(processed_texts)

X = pad_sequences(sequences, maxlen=5)
y = np.array(labels)

model = Sequential()
model.add(Embedding(input_dim=100, output_dim=16, input_length=5))
model.add(LSTM(32))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X, y, epochs=10)

test_text = "I feel very stressed"
doc = nlp(test_text)
test_processed = " ".join([token.lemma_.lower() for token in doc])

seq = tokenizer.texts_to_sequences([test_processed])
pad = pad_sequences(seq, maxlen=5)

prediction = model.predict(pad)

print("User Input:", test_text)
print("Prediction:", prediction)
```
